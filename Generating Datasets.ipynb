```python
from faker import Faker
fake = Faker()
```

```python
from random import randint 
import pandas as pd 
```

# Creating a dataframe with synthetic data

```python
def input_data(x):
   
    # pandas dataframe
    data = pd.DataFrame()
    for i in range(0, x):
        data.loc[i,'id']= randint(10000, 99999)
        data.loc[i,'name']= fake.name()
        data.loc[i,'age']= randint(18, 89)
        data.loc[i,'address']= fake.address()
    return data
   

input_data(20)
```
