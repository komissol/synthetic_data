```python
from faker import Faker
#Creating instance of faker
fake = Faker()
```

```python
from random import randint 
import pandas as pd 
```

```python
# Creating a dataframe with synthetic data
def input_data(x):
   
    # pandas dataframe
    data = pd.DataFrame()
    for i in range(0, x):
        data.loc[i,'id']= randint(10000, 99999)
        data.loc[i,'name']= fake.name()
        data.loc[i,'age']= randint(18, 89)
        data.loc[i,'address']= fake.address()
    return data
   

input_data(20)
```

```python
import datetime
import pandas as pd
import numpy as np
```

```python
day = 24 * 60 * 60
year = 365.2425 * day
```

```python
def load_dataframe() -> pd.DataFrame:
    """ Create a time series x sin wave dataframe. """
    df = pd.DataFrame(columns=['date', 'sin'])
    df.date = pd.date_range(start='2019-01-01', end='2023-03-01', freq='D')
    df.sin = 1 + np.sin(df.date.astype('int64') // 1e9 * (2 * np.pi / year))
    df.sin = (df.sin * 100).round(2)
    df.date = df.date.apply(lambda d: d.strftime('%Y-%m-%d'))
    return df

train_df = load_dataframe()
```

# 2. Generating Times-Series Data (basic)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
DATE_START = '2019-01-01'
DATE_END = '2019-12-31'
dates = pd.date_range(DATE_START, DATE_END)
df = pd.DataFrame({
    'date': dates,
    'value': np.random.normal(0,1,dates.size)
})
df.set_index('date', inplace=True)
plt.plot(df['value'])
plt.ylabel('Value')
plt.xlabel('Date')
plt.title('Random Values')
plt.show()
```
